```python
# đọc thử luôn file raw (170MB)
# TODO: sau này chuyển qua dùng sample
df = pd.read_csv("../data/raw/6_arXiv_scientific_dataset.csv")
df.head()
```

```python
# check sơ cấu trúc
df.shape, df.columns
```

```python
# thử lấy 5 dòng random xem sao
df.sample(5)
```

```python
# lọc mấy paper sau 2018
recent = df[df["year"] >= 2018]
recent.shape
```

```python
# TODO: convert cột ngày tháng, chưa chắc tên cột đúng
# df["submitted_date"] = pd.to_datetime(df["submitted"], errors="coerce")
# df["year"] = df["submitted_date"].dt.year
# df["month"] = df["submitted_date"].dt.month
#
# df[["submitted", "submitted_date", "year", "month"]].head()
```

```python
# check lại phân bố năm (dùng recent hay full? để tạm vậy)
recent["year"].value_counts().sort_index().tail(10)
```

```python
# vẽ thử histogram, để tạm hard-code luôn
plt.figure(figsize=(8,4))
df["year"].hist(bins=30)
plt.title("Year distribution (all)")
plt.show()
```

```python
# THỬ ý tưởng: chỉ lấy các paper thuộc AI/ML
# NOTE: cột category hình như là 'categories' hay 'category', để thử sau
ai_mask = df["categories"].str.contains("cs.AI", na=False) | df["categories"].str.contains("cs.LG", na=False)
df_ai = df[ai_mask]
df_ai.shape
```

```python
# lấy sample nhỏ để test idea, sau này sẽ chuyển ra script riêng
df_ai_sample = df_ai.sample(5000)
df_ai_sample.head(3)
```

```python
# thử tính top authors, làm nhanh cho rồi
# FIXME: tách author đang hơi ẩu, chỉ split dấu phẩy
df_ai_sample["first_author"] = df_ai_sample["authors"].str.split(",").str[0]
df_ai_sample["first_author"].value_counts().head(10)
```

```python
# TODO: đưa phần này vào function sau
def plot_year_counts(tmp_df, title):
    tmp_df["year"].value_counts().sort_index().plot(kind="bar", figsize=(10,4))
    plt.title(title)
    plt.tight_layout()
    plt.show()

plot_year_counts(df_ai_sample, "AI sample year distribution")
```

```python
# thử một sample khác xem shape có giống không
df_ai_sample2 = df_ai.sample(5000)
df_ai_sample2["year"].value_counts().sort_index().tail()
```

```python
# IDEA: so sánh 2 sample, tạm làm nhanh
pd.concat(
    [
        df_ai_sample["year"].value_counts().sort_index().rename("sample1"),
        df_ai_sample2["year"].value_counts().sort_index().rename("sample2")
    ],
    axis=1
).head(10)
```

```python
# tạm thời lưu sample này ra file cho notebook khác dùng
df_ai_sample.to_csv("../data/processed/arxiv_ai_sample_temp.csv", index=False)
```
